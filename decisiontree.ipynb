```r
a = c(1,2,3,1,2,3) 
b = c(3,3,1,1,1,1) 
c= c(2,1,1,2,5,2)
d=c("a","a","a","b","b","b")
df = data.frame(a,b,c,d)
```

```r
library(rpart)
```

```r
fit <- rpart(df$d ~ a + b + c, method="class", data=df)
```

```r
plot(fit, uniform=TRUE, main="Classification Tree for Toy Data", control =rpart.control(minsplit =1,minbucket=1, cp=1))
```

```r
summary(fit)
plotcp(fit)
```
